```julia
###############################
# Lighting Notes Generator (Version 2)
###############################

# --- FULL General Notes ---
general_notes = [
    "1. ALL OUTLETS WITHIN 6' OF A SINK SHALL BE GFCI PROTECTED. -1",
    "2. ALL GFCI OUTLETS SHALL BE INSTALLED IN A READILY ACCESSIBLE LOCATION PER NEC 210.8. IF EQUIPMENT IS TO BE INSTALLED IN FRONT OF A GFCI OUTLET PROVIDE DEAD FRONT GFCI IN READILY ACCESSIBLE LOCATION FEEDING A STANDARD OUTLET AND LABEL THE GFCI DEADFRONT WITH THE EQUIPMENT NAME. -1",
    "3. COORDINATE LOCATION AND MOUNTING HEIGHT OF DEVICES WITH ARCHITECTURAL AND FOOD SERVICE FLOOR PLANS AND ELEVATIONS PRIOR TO ROUGH-IN.",
    "4. COORDINATE ALL EQUIPMENT REQUIREMENTS WITH MANUFACTURER PRIOR TO ROUGH IN.",
    "5. CONTRACTOR SHALL COORDINATE A SITE VISIT WITH ARCHITECT AND OWNER TO REVIEW LIGHTING, OUTLETS, SWITCHES, TELEPHONE, DATA, T.V. ET. HARDWARE LOCATIONS PRIOR TO INSTALLATION.",
    "6. ALL CONDUIT SHALL BE CONCEALED. IN EXPOSED CEILING AREAS, EXPOSED AND SURFACE MOUNTED CONDUIT SHALL BE EMT AND SHALL BE ROUTED ORTHOGONAL TO THE STRUCTURE AND SHALL BE INSTALLED IN NEAT AND WORKMAN LIKE MANNER.",
    "7. ALL DEVICES INCLUDING OUTLETS AND SWITCHES SHALL BE WHITE FINISH WITH WHITE COVER PLATES. COORDINATE ALL FINISHES WITH ARCHITECT.",
    "8. INSTALL ALL DEVICES PER ADA REQUIREMENTS.",
    "9. ALL BRANCH CIRCUITS OVER 50' SHALL USE #10 CONDUCTORS.",
    "10. LABEL ALL OUTLETS WITH PANEL NAME AND CIRCUIT NUMBER USING CLEAR LABEL WITH BLACK PRINTED LETTERS.",
    "11. PROVIDE HANDLE TIES FOR SHARED NEUTRAL CIRCUIT BREAKERS IN PANELBOARDS.",
    "12. ALL PATIENT CARE AREAS SHALL COMPLY WITH NEC SECTION 517.13(A) AND 517.13(B). ALL PATIENT CARE RECEPTACLES AND FIXED EQUIPMENT SHALL BE GROUNDED BY AN INSULATED COPPER CONDUCTOR SIZE PER TABLE 250.122. IN ADDITION THE CIRCUIT SERVING THE PATIENT CARE RECEPTACLES AND FIXED EQUIPMENT SHALL BE INSTALLED IN A METAL RACEWAY OR CABLE THAT QUALIFIES AS AN EQUIPMENT GROUNDING RETURN PATH IN ACCORDANCE WITH SECTION 250.118. -2",
    "13. MULTI-WIRE BRANCH CIRCUITS SUPPLYING POWER TO THE SYSTEM FURNITURE SHALL BE PROVIDED WITH A MEANS TO DISCONNECT SIMULTANEOUSLY ALL UNGROUNDED CONDUCTORS AT POWER PANEL BOARD WHERE THE BRANCH CIRCUITS ORIGINATE. -3 or 4"
]

# --- FULL Key Notes ---
key_notes = [
    "1. SYSTEMS FURNITURE WALL FEED. PROVIDE CONNECTION FOR POWER AND TELE/DATA. -3",
    "2. SYSTEMS FURNITURE FLOOR FEED. PROVIDE CONNECTION FOR POWER AND TELE/DATA. -4",
    "3. PROVIDE NEW FLOOR CORE OUTLET. COORDINATE LOCATION WITH TENANT AND FURNITURE VENDOR. -4",
    "4. MECHO SHADE. PROVIDE SWITCH, J-BOX AND EQUIPMENT CONNECTION. VERIFY REQUIREMENTS WITH MANUFACTURER. -6",
    "5. MOTOR RATED SWITCH FOR EXHAUST FAN. MOUNT UNDER COMMON COVER PLATE WITH LIGHT SWITCHES. VERIFY LOCATION OF EXHAUST FAN WITH MECHANICAL. -6",
    "6. J-BOX FOR WATER COOLER. VERIFY CONNECTION WITH MECHANICAL. -6",
    "7. J-BOX FOR MOTORIZED PROJECTION SCREEN. PROVIDE UP/DOWN SWITCH. VERIFY REQUIREMENTS AND MOUNTING WITH ARCHITECT. -6",
    "8. J-BOX FOR CONTROL POWER TO FLUSH VALVE. COORDINATE LOCATION WITH MECHANICAL AND MAKE FINAL CONNECTION TO CONTROL TRANSFORMER FURNISHED BY MECHANICAL. -6",
    "9. J-BOX FOR CONTROL POWER TO VAV. COORDINATE LOCATION WITH MECHANICAL AND MAKE FINAL CONNECTION TO CONTROL TRANSFORMER FURNISHED BY MECHANICAL. -6",
    "10. PROVIDE J-BOX AND CONNECTION FOR INSTA-HOT WATER HEATER. VERIFY REQUIREMENTS WITH MECHANICAL. -1",
    "11. PROVIDE WALL MOUNTED SWITCH TO CONTROL EXHAUST FAN. -6",
    "12. PROVIDE 120V RECEPTACLE FOR CONDENSATE PUMP. COORDINATE LOCATION AND REQUIREMENTS WITH MECHANICAL CONTRACTOR. -6",
    "13. PROVIDE DEDICATED CIRCUIT FOR ELEVATOR SMOKE GUARD DOOR. PROVIDE LOCKABLE RED CIRCUIT BREAKER IN PANELBOARD. -6",
    "14. PROVIDE DEDICATED CIRCUIT FOR WON DOOR. PROVIDE LOCKABLE RED CIRCUIT BREAKER IN PANELBOARD. -6",
    "15. PROVIDE CONNECTION TO SMOKE FIRE DAMPER. VERIFY CONNECTION REQUIREMENTS WITH MECHANICAL. PROVIDE LOCKABLE RED CIRCUIT BREAKER IN PANELBOARD. -6"
]


###########################################
# USER INPUT
###########################################

println("Enter 1 or 0 for each condition.")

print("Is there a sink? (1/0): ")
sink = parse(Int, readline())

print("Is there patient care? (1/0): ")
patient = parse(Int, readline())

print("Is there a wall furniture feed? (1/0): ")
wall_feed = parse(Int, readline())

print("Is there a floor furniture feed? (1/0): ")
floor_feed = parse(Int, readline())


###########################################
# SELECT NOTES
###########################################

general_out = String[]
key_out = String[]
misc_out = String[]  # for all -6 notes

# ---- Process General Notes ----
for note in general_notes
    if endswith(note, "-1") && sink == 1
        push!(general_out, note)
    elseif endswith(note, "-2") && patient == 1
        push!(general_out, note)
    elseif occursin("-3 or 4", note) && (wall_feed == 1 || floor_feed == 1)
        push!(general_out, note)
    elseif !(endswith(note, "-1") || endswith(note, "-2") || occursin("-3 or 4", note))
        push!(general_out, note)
    end
end

# ---- Process Key Notes ----
for note in key_notes
    if endswith(note, "-3") && wall_feed == 1
        push!(key_out, note)
    elseif endswith(note, "-4") && floor_feed == 1
        push!(key_out, note)
    elseif endswith(note, "-1") && sink == 1
        push!(key_out, note)
    elseif endswith(note, "-6")
        push!(misc_out, note)
    elseif !(endswith(note, "-1") || endswith(note, "-3") || endswith(note, "-4") || endswith(note, "-6"))
        push!(key_out, note)
    end
end


###########################################
# Strip Contingency Tags (-1, -2, -3, -4, -6)
###########################################

function strip_tags(notes::Vector{String})
    return [replace(n, r"\s*-(1|2|3|4|6)(\s*or\s*4)?$" => "") for n in notes]
end

general_out = strip_tags(general_out)
key_out = strip_tags(key_out)
misc_out = strip_tags(misc_out)


###########################################
# Renumber Notes Sequentially
###########################################

function renumber(notes::Vector{String})
    newlist = String[]
    for (i, note) in enumerate(notes)
        cleaned = replace(note, r"^\d+\.\s*" => "")
        push!(newlist, string(i, ". ", cleaned))
    end
    return newlist
end

general_out = renumber(general_out)
key_out = renumber(key_out)
misc_out = renumber(misc_out)


###########################################
# OUTPUT
###########################################

println("\n=============================")
println("APPLICABLE NOTES")
println("=============================\n")

println("GENERAL NOTES:")
for n in general_out
    println(n)
end

println("\nKEY NOTES:")
for n in key_out
    println(n)
end

println("\nMISCELLANEOUS RARE NOTES:")
for n in misc_out
    println(n)
end
```
