```julia
#############################################
# Demolition Notes Generator
#############################################

general_notes = [
    "1. MAINTAIN CONTINUITY OF CIRCUIT TO EXISTING SPACES OUTSIDE OF THE DEMOLITION SCOPE OF WORK. RE-ROUTE, INTERCEPT AND EXTEND CONDUIT AND CONDUCTORS AS NECESSARY.",
    "2. REFER TO ARCHITECTURAL DEMOLITION PLANS FOR DEMOLITION SCOPE OF WORK.",
    "3. FIELD VERIFY EXISTING BRANCH CIRCUITRY PRIOR TO COMMENCEMENT OF WORK.",
    "4. EXAMINE PANELBOARDS AND DISTRIBUTION EQUIPMENT AFFECTED BY THE WORK PRIOR TO DEMOLITION.",
    "5. OUTLETS AND DEVICES DENOTED WITH AN 'X' ARE EXISTING TO REMAIN, MAINTAIN CIRCUITRY AS REQUIRED FOR DEMOLITION. -1",
    "6. OUTLETS AND DEVICES DENOTED WITH AN 'XR' ARE EXISTING TO BE REMOVED, MAINTAIN CIRCUITRY TO OTHER DEVICES AS REQUIRED FOR DEMOLITION."
]

key_notes = [
    "1. DEMOLISH ALL LIGHT FIXTURES, CONTROLS, WIRE, CONDUIT AND SUPPORTS BACK TO SOURCE THROUGHOUT.",
    "2. DEMOLISH ALL PANELBOARDS, OUTLETS AND JUNCTION BOXES, WIRE, CONDUIT AND SUPPORTS BACK TO SOURCE THROUGHOUT."
]


#############################################
# USER INPUT
#############################################

println("Are there existing outlets/devices (meaning -1 notes apply)? Enter 1 for yes, 0 for no.")
print("Are there existing outlets/devices? (1/0): ")
existing_outlets = parse(Int, readline())


#############################################
# FILTER NOTES BASED ON CONTINGENCY
#############################################

general_out = String[]
key_out = String[]

for note in general_notes
    if endswith(note, "-1")
        if existing_outlets == 1
            push!(general_out, note)
        end
    else
        push!(general_out, note)
    end
end

key_out = key_notes  # Always included


#############################################
# STRIP TAGS (-1)
#############################################

function strip_tags(notes::Vector{String})
    return [replace(n, r"\s*-(1)$" => "") for n in notes]
end

general_out = strip_tags(general_out)
key_out = strip_tags(key_out)


#############################################
# RENUMBER SEQUENTIALLY
#############################################

function renumber(notes::Vector{String})
    newlist = String[]
    for (i, note) in enumerate(notes)
        cleaned = replace(note, r"^\d+\.\s*" => "")
        push!(newlist, string(i, ". ", cleaned))
    end
    return newlist
end

general_out = renumber(general_out)
key_out = renumber(key_out)


#############################################
# OUTPUT WITH SPACING BETWEEN NOTES
#############################################

println("\n=============================")
println("APPLICABLE DEMOLITION NOTES")
println("=============================\n")

println("GENERAL NOTES:")
for n in general_out
    println(n)
    println()   # <-- adds one blank line between notes
end

println("KEY NOTES:")
for n in key_out
    println(n)
    println()   # <-- adds one blank line between notes
end
```
