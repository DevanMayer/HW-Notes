```julia
###############################
# Lighting Notes Generator
###############################

# --- FULL General Notes ---
general_notes = [
    "1. COORDINATE LOCATION AND MOUNTING HEIGHT OF DEVICES AND LIGHT FIXTURES WITH ARCHITECTURAL FLOOR PLANS, REFLECTED CEILING PLANS AND ELEVATIONS.",
    "2. COORDINATE LIGHT FIXTURE INSTALLATION HARDWARE WITH CEILING TYPES.",
    "3. CONTRACTOR SHALL COORDINATE A SITE VISIT WITH ARCHITECT AND OWNER TO REVIEW LIGHTING, OUTLETS, SWITCHES, TELEPHONE, DATA, T.V. ET. HARDWARE LOCATIONS PRIOR TO INSTALLATION.",
    "4. ALL CONDUIT SHALL BE CONCEALED. IN FINAL (NON TEMPORARY) EXPOSED CEILING AREAS, EXPOSED AND SURFACE MOUNTED CONDUIT SHALL BE EMT AND SHALL BE ROUTED ORTHOGONAL TO THE STRUCTURE AND SHALL BE INSTALLED IN NEAT AND WORKMAN LIKE MANNER.",
    "5. ALL DEVICES INCLUDING OUTLETS AND SWITCHES TO BE DECORA WITH WHITE FINISH AND WHITE COVER PLATES. COORDINATE ALL FINISHES WITH ARCHITECT.",
    "6. INSTALL ALL LIGHT SWITCHES UNDER A COMMON DEVICE PLATE WHERE GROUPED TOGETHER.",
    "7. ALL WIRE ROUTING SHOWN IS DIAGRAMMATIC, CONTRACTOR TO COORDINATE FINAL ROUTING WITH FIELD CONDITIONS.",
    "8. COORDINATE OCCUPANCY AND DAYLIGHT SENSOR PLACEMENT WITH MANUFACTURES RECOMMENDATIONS.",
    "9. MOUNT ALL DEVICES PER ADA REQUIREMENTS.",
    "10. PROVIDE HOT UNSWITCHED LEG TO EMERGENCY BATTERY PACK. -1"
]

# --- FULL Key Notes ---
key_notes = [
    "1. LESS THAN 120 WATTS OF GENERAL LIGHTING IN THE PRIMARY DAY-LIT ZONE AND LESS THAN 120 WATTS OF GENERAL LIGHTING IN THE SECONDARY DAY-LIT ZONE. DAYLIGHTING CONTROLS NOT REQUIRED PER EXCEPTION 3 TO 2022 BEES SECTION 130.1(D).",
    "2. PROVIDE HOT UNSWITCHED LEG TO EMERGENCY BATTERY PACK. -1",
    "3. PROVIDE NEW EXIT SIGN, CONNECT CIRCUIT VIA LIGHTING INVERTER. -1",
    "4. PROVIDE NEW EXIT SIGN, CONNECT CIRCUIT TO BUILDING EMERGENCY GENERATOR. -2",
    "5. LIGHT SWITCHES ARE TO BE LOCATED IN THE ### ROOM. PER EXCEPTION TO SECTION 130.1(a)1: AREAS OF THE BUILDING INTENDED FOR ACCESS OR USE BY THE PUBLIC MAY USE A MANUAL CONTROL NOT ACCESSIBLE TO UNAUTHORIZED PERSONNEL. -3",
    "6. OPEN OFFICE AREAS TO BE BROKEN INTO SEPARATE DIMMING ZONES NO GREATER THAN 600 SQFT FOR COMPLIANCE WITH 2022 BEES SECTION 130.1(c)6D. -4"
]

println("Enter 1 or 0 for each condition:")

# --- Emergency Input ---
print("Emergency battery system present? (1/0): ")
emergency = parse(Int, readline())

# Auto-opposite
not_emergency = emergency == 1 ? 0 : 1

print("Important public areas present? (1/0): ")
public_areas = parse(Int, readline())

print("Is there an open office area? (1/0): ")
open_office = parse(Int, readline())

###############################
# Determine Applicable Notes
###############################
general_out = String[]
key_out = String[]

# --- General Notes (contingency: -1) ---
for note in general_notes
    if endswith(note, "-1")
        if emergency == 1
            push!(general_out, note)
        end
    else
        push!(general_out, note)
    end
end

# --- Key Notes ---
for note in key_notes
    if endswith(note, "-1") && emergency == 1
        push!(key_out, note)
    elseif endswith(note, "-2") && not_emergency == 1
        push!(key_out, note)
    elseif endswith(note, "-3") && public_areas == 1
        push!(key_out, note)
    elseif endswith(note, "-4") && open_office == 1
        push!(key_out, note)
    elseif !(endswith(note, "-1") || endswith(note, "-2") || endswith(note, "-3") || endswith(note, "-4"))
        push!(key_out, note)
    end
end

###############################
# Remove Contingency Tags (-1/-2/-3/-4)
###############################
function strip_tags(notes::Vector{String})
    return [replace(n, r"\s*-([1-4])$" => "") for n in notes]
end

general_out = strip_tags(general_out)
key_out = strip_tags(key_out)

###############################
# Automatic Renumbering
###############################
function renumber(notes::Vector{String})
    new_notes = String[]
    for (i, note) in enumerate(notes)
        cleaned = replace(note, r"^\d+\.\s*" => "")
        push!(new_notes, string(i, ". ", cleaned))
    end
    return new_notes
end

general_out = renumber(general_out)
key_out = renumber(key_out)

###############################
# OUTPUT WITH SPACING BETWEEN NOTES
###############################
println("\n=============================")
println("APPLICABLE NOTES")
println("=============================\n")

println("GENERAL NOTES:")
for n in general_out
    println(n)
    println()   # <-- adds blank line between notes
end

println("KEY NOTES:")
for n in key_out
    println(n)
    println()   # <-- adds blank line between notes
end
```
